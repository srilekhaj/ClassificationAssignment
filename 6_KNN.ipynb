```python
#importing the Libraies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
#input output split
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 
                       'rbc_normal', 'pc_normal', 'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[["classification_yes"]]
```

```python
independent.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
# standaridation value giving to model
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)   #z = (x - u) / s
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}



grid = GridSearchCV(KNeighborsClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid.fit(X_train, y_train) 
 

```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred)
```

```python
print(clf_report) #without standardization 
```

```python
print(cm)
```

```python
print("KNN Classification Report")
print(clf_report) #with standardization 
```

```python
from sklearn.metrics import roc_auc_score
# Assuming grid_search is your GridSearchCV object
best_model = grid.best_estimator_

# Now you can call predict_proba on the best model
#probabilities = best_model.predict_proba(X_test)  # X_test is your test data

roc_auc_score(y_test,best_model.predict_proba(X_test)[:,1])  #you should give probability=true in gridsearch to avoid error
```
