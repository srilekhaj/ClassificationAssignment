```python
#importing the Libraies
import numpy as np
# import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True, dtype=int)
```

```python
dataset["classification_yes"].value_counts()
```

```python
#input output split
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 
                       'rbc_normal', 'pc_normal', 'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[["classification_yes"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
```

```python
def Gaussian():
    classifier = GaussianNB()
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    
    cm = confusion_matrix(y_test, y_pred)
   
    clf_report = classification_report(y_test, y_pred)

    roc_score = roc_auc_score(dependent,classifier.predict_proba(independent)[:,1])
    print(cm)
    print(clf_report)
    print(roc_score)
```

```python
def roc_score():
    from sklearn.metrics import roc_auc_score
    # Assuming grid_search is your GridSearchCV object
    # best_model = classifier.best_estimator_
    
    # Now you can call predict_proba on the best model
    #probabilities = best_model.predict_proba(X_test)  # X_test is your test data
    
    return roc_auc_score(dependent,classifier.predict_proba(independent)[:,1])  #you should give probability=true in gridsearch to avoid error
```

```python
def Multinomial():    
    classifier = MultinomialNB()
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    clf_report = classification_report(y_test, y_pred)
    roc_score = roc_auc_score(dependent,classifier.predict_proba(independent)[:,1])
    print(cm)
    print(clf_report)
    print(roc_score)
```
