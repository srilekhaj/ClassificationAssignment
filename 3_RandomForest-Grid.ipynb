```python
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True, dtype=int)
dataset
```

```python
dataset['classification_yes'].value_counts()
```

```python
dataset.columns
```

```python
#input output split
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 
                       'rbc_normal', 'pc_normal', 'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[["classification_yes"]]
```

```python
#train test split'
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier


param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_features' : ['sqrt', 'log2'], 
    'n_estimators' : [10, 100]
} 



grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid.fit(X_train, y_train) 
 

```

```python
grid_predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
```

```python
print(cm)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
print("Random Forest Classification Report")
print(clf_report)
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(y_test,grid_predictions)
print(a)
```

```python
from sklearn.metrics import f1_score
a = f1_score(y_test,grid_predictions, average='weighted')  #bydefault give class 1 f1score 
print(a)
```

```python
print(grid.best_params_)
```

```python
from sklearn.metrics import roc_auc_score
# Assuming grid_search is your GridSearchCV object
best_model = grid.best_estimator_

# Now you can call predict_proba on the best model
#probabilities = best_model.predict_proba(X_test)  # X_test is your test data

roc_auc_score(y_test,best_model.predict_proba(X_test)[:,1])  #you should give probability=true in gridsearch to avoid error
```
