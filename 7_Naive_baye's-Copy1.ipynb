```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True, dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
#input output split
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 
                       'rbc_normal', 'pc_normal', 'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[["classification_yes"]]
```

```python
independent.shape
```

```python
dependent.shape
```

```python
from sklearn.naive_bayes import CategoricalNB
classifier =CategoricalNB()
classifier.fit(independent, dependent)
```

```python
y_pred = classifier.predict(dependent)
```

```python

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(dependent, y_pred)
from sklearn.metrics import classification_report
clf_report = classification_report(dependent, y_pred)
print(clf_report)
print(cm)
```
